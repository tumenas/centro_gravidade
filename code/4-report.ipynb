### 1- Introdução

A decisão sobre a localização de instalações é um dos principais problemas na área de operações. A localização tem impacto direto na velocidade e no custo para uma operação atender seu público-alvo. Uma localização mais pŕoxima do mercado consumidor pode ser mais veloz no atendimento aos clientes. No entanto, quase certamente terá um custo maior do que uma localização mais afastada. 
Uma das ferramentas disponíveis para escolha da localização é o método do centro de gravidade. O método do centro de gravidade é uma abordagem quantitativa que minimiza o custo total de transporte entre o ponto escolhido e a localização das origens ou destinos. O método do centro de gravidade tem sido usado em várias situações (Bittencourt; Junior, 2021; De Sousa et al, 2019; De Sousa Pereira et al, 2019; Pamplona; De Castro Fortes;  Alves, 2014; Rodrigues et al, 2023; Sellitto et al, 2015; Soares et al, 2020), mas uma das aplicações mais famosas foi a realizada pela empresa norte-americana de entregas Federal Express (FedEx). A estratégia inicial da FedEx era transportar todas as entregas para um ponto (chamado de hub logístico) durante a tarde e a noite, classificá-las nesse hub logístico e, em seguida, transportá-las para seus destinos durante a noite para entrega no dia seguinte. O problema ficou conhecido como *“FedEx Problem”* (Morrison, 2010), devido à decisão da FedEx de estabelecer seu hub logístico na cidade de Memphis. Memphis foi escolhida como solução do seguinte problema: Dada uma população representada por pontos no espaço euclidiano, qual é a localização que minimiza a distância total a todos os pontos? Essa localização é o centro de gravidade.

Apesar do método do centro de gravidade ser uma abordagem tradicional, com diversas aplicações em diversos setores, não há informações na literatura sobre a localização do centro de gravidade do Brasil. Nos EUA o centro de gravidade do país reflete a localização de  operações logísticas importantes, como o caso da FedEx. Será que, no centro de gravidade do Brasil, também existem operações logísticas de relevância? 

Este trabalho busca responder qual o centro de gravidade do Brasil, assim como quais são os centros de gravidade das cinco regiões brasileiras (Norte, Nordeste, Centro-Oeste, Sudeste e Sul), utilizando os dados mais recentes da população brasileira do Censo de 2022.


### 2- Revisão Teórica ###

Na logística, o método do centro de gravidade é utilizado para derivar a localização ideal para instalações de armazenamento, calculando os custos de transporte para cada um dos locais de oferta ou demanda a serem atendidos pela instalação. O centro de gravidade é o local que apresenta os custos totais de transporte mais baixos.

Pelo seu nome, o método do centro de gravidade é análogo ao cálculo do centro de massa, com a substituição da massa pelos volumes a serem transportados. A fórmula para as coordenadas x e y do centro de gravidade (ou do centro de massa) é dada por

$\
x=\frac{\sum_{i=1}^{n}x_{i}m_{i}}{\sum_{i=1}^{n}m_{i}}
\\
y=\frac{\sum_{i=1}^{n}y_{i}m_{i}}{\sum_{i=1}^{n}m_{i}}
$

Onde x e y são coordenadas dos “n” locais atendidos (coordenadas dos “n” pontos de massa distribuídos) e ‘m’ refere-se aos volumes a serem transportados (massa para o centro de massa físico). 

Outra maneira é encarar o centro de gravidade como um ponto no espaço onde a localização de seus eixos x e y (latitude e longitude, respectivamente) é calculada por meio da média ponderada pelos volumes (ou massa) de todos os pontos existentes no espaço analisado. Trata-se de uma técnica simples para a escolha do local de localização de uma localização. No caso do centro de gravidade populacional, o valor de ‘m’ será o tamanho da população de cada localidade da região analisada.

os tomadores de decisões das empresas ainda não utilizam métodos científicos na hora de escolher a localização de suas instalações (Sato, 2002)
