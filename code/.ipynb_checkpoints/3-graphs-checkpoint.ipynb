```python
#https://ipeagit.github.io/geobr/articles/python-intro/py-intro-to-geobr.html

import geobr
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
import warnings
warnings.filterwarnings('ignore')


%matplotlib inline

# Centro gravitacional
lat = -16.990485398957283 # Y
long = -45.8020766490012 # X
```

```python
state = geobr.read_state(code_state="SE", year=2018)
states = geobr.read_state(year=2019)
```

```python
# Gráfico do Brasil
fig, ax = plt.subplots(figsize=(15, 15), dpi=300)

states.plot(facecolor="white", edgecolor="black", ax=ax)

# Centro gravitacional
lat = -16.990485398957283 # Y
long = -45.8020766490012 # X

#Brasilândia de Minas, MG
plt.scatter(long, lat, color='red', s=100) 
plt.text(long - 1, lat - 1, 'Centro de Gravidade', fontsize=12, color = 'red')

ax.set_title("Mapa do Brasil", fontsize=20)
ax.axis("off")

plt.savefig('/home/tumenas/Felipe/Papers/centro_gravidade/results/brasil.png')
```

```python
# Gráfico das regiões
states = geobr.read_state()
gdf = gpd.GeoDataFrame(states)

no_states = gdf[gdf["name_region"] == "Norte"]
ne_states = gdf[gdf["name_region"] == "Nordeste"]
co_states = gdf[gdf["name_region"] == "Centro Oeste"]
se_states = gdf[gdf["name_region"] == "Sudeste"]
su_states = gdf[gdf["name_region"] == "Sul"]
```

```python
# Gráfico da Região Norte
fig, ax = plt.subplots(figsize=(15, 15), dpi=300)

no_states.plot(facecolor="white", edgecolor="black", ax=ax)


# Centro gravitacional
lat = -4.163864 # Y
long = -54.762229 # X

# Cidade : Rurópolis, PA

plt.scatter(long, lat, color='red', s=100) 
plt.text(long - 1, lat - 1, 'Centro de Gravidade', fontsize=12, color = 'red')

#ax.set_title("Mapa do Brasil", fontsize=20)
ax.axis("off")

plt.savefig('/home/tumenas/Felipe/Papers/centro_gravidade/results/no.png')
```

```python
# Gráfico da Região Nordeste

fig, ax = plt.subplots(figsize=(15, 15), dpi=300)

ne_states.plot(facecolor="white", edgecolor="black", ax=ax)

# Centro gravitacional
lat = -8.120984 # Y
long = -39.036280 # X

# Cidade : Salgueiro, PE

plt.scatter(long, lat, color='red', s=100) 
plt.text(long - 0, lat - 0.5, 'Centro de Gravidade', fontsize=12, color = 'red')

#ax.set_title("Mapa do Brasil", fontsize=20)
ax.axis("off")

plt.savefig('/home/tumenas/Felipe/Papers/centro_gravidade/results/ne.png')
```
